```python
1. Add the current date to the text file today.txt as a string.
Ans:-
import datetime
#  Add current date to the today.txt file
file = open('today.txt','w')
file.write(datetime.datetime.now().strftime("%d-%m-%Y"))
file.close()
#  Read current date from today.txt file
file = open('today.txt','r')
print(file.read())
file.close()

Output:-
     05-08-2022
        
2. Read the text file today.txt into the string today_string
Ans:-
file = open('today.txt','r')
today_string = file.read()
print(today_string)

Output:-
    05-08-2022
    
3. Parse the date from today_string.
Ans:-
from datetime import datetime
parsed_data = datetime.strptime(today_string, '%d-%m-%Y')
print(parsed_data)

Output:-
    2022-08-05 00:00:00
            
4. List the files in your current directory
Ans:-
import os
for folders, subfolders, files in os.walk(os.getcwd()):
    for file in files:
        print(file)
        
Output:-
    .condarc
123.ipynb
18th Dec.ipynb
18th Dem 2021.ipynb
19 Dec python program Discussion .ipynb
19th dec.ipynb
4th Dec.ipynb
5th Dec.ipynb
abc.csv
Assigement_12.ipynb
Assigment.7.ipynb
Assigment.8 .ipynb
Assigment.9.ipynb
Assigment_16.ipynb
Assigment_17.ipynb
Assigment_18.ipynb
Assigment_19.ipynb
Assigment_4.ipynb
Assignment-10.ipynb
Assignment-9.ipynb
Assignment.3.ipynb
Assignment.5.ipynb
Assignment_11.ipynb
Assignment_13.ipynb
Assignment_14.ipynb
Assignment_20.ipynb
Assignment_21.ipynb
Assignment_6.ipynb
Assignment__15.ipynb
books.csv
books.db
Contents.txt
Download.ipynb
Graph.ipynb
ineuron.db
jobdb.db
NTUSER.DAT
ntuser.dat.LOG1
ntuser.dat.LOG2
NTUSER.DAT{53b39e88-18c4-11ea-a811-000d3aa4692b}.TM.blf
NTUSER.DAT{53b39e88-18c4-11ea-a811-000d3aa4692b}.TMContainer00000000000000000001.regtrans-ms
NTUSER.DAT{53b39e88-18c4-11ea-a811-000d3aa4692b}.TMContainer00000000000000000002.regtrans-ms
ntuser.ini
pandas and numpy.ipynb
Pandas.ipynb
PCA Implemention.ipynb
persons.csv
Python Assignment _2.ipynb
Python basic (3).ipynb
python first proj.....ipynb
Python-Assigment-4.ipynb
test.txt
today.txt
Untitled.ipynb
Untitled1.ipynb
Untitled10.ipynb
Untitled11.ipynb
Untitled12.ipynb
Untitled13.ipynb
Untitled14.ipynb
Untitled15.ipynb
Untitled2.ipynb
Untitled3.ipynb
Untitled4.ipynb
Untitled5.ipynb
Untitled6.ipynb
Untitled7.ipynb
Untitled8.ipynb
Untitled9.ipynb
zoo.ipynb
environments.txt
config.yaml
.config
123-checkpoint.ipynb
18th Dec-checkpoint.ipynb
18th Dem 2021-checkpoint.ipynb
19 Dec python program Discussion -checkpoint.ipynb
19th dec-checkpoint.ipynb
4th Dec-checkpoint.ipynb
5th Dec-checkpoint.ipynb
Assigement_12-checkpoint.ipynb
Assigment.7-checkpoint.ipynb
Assigment.8 -checkpoint.ipynb
Assigment.9-checkpoint.ipynb
Assigment_16-checkpoint.ipynb
Assigment_17-checkpoint.ipynb
Assigment_18-checkpoint.ipynb
Assigment_19-checkpoint.ipynb
Assigment_4-checkpoint.ipynb
Assignment-10-checkpoint.ipynb
Assignment-9-checkpoint.ipynb
Assignment.3-checkpoint.ipynb
Assignment.5-checkpoint.ipynb
Assignment_11-checkpoint.ipynb
Assignment_13-checkpoint.ipynb
Assignment_14-checkpoint.ipynb
Assignment_20-checkpoint.ipynb
Assignment_21-checkpoint.ipynb
Assignment_6-checkpoint.ipynb
Assignment__15-checkpoint.ipynb
Download-checkpoint.ipynb
Graph-checkpoint.ipynb
pandas and numpy-checkpoint.ipynb
Pandas-checkpoint.ipynb
PCA Implemention-checkpoint.ipynb
Python Assignment _2-checkpoint.ipynb
Python basic (3)-checkpoint.ipynb
python first proj....-checkpoint.ipynb
Python-Assigment-4-checkpoint.ipynb
Untitled-checkpoint.ipynb
Untitled1-checkpoint.ipynb
Untitled10-checkpoint.ipynb
Untitled11-checkpoint.ipynb
Untitled12-checkpoint.ipynb
Untitled13-checkpoint.ipynb
Untitled14-checkpoint.ipynb
Untitled15-checkpoint.ipynb
Untitled2-checkpoint.ipynb
Untitled3-checkpoint.ipynb
Untitled4-checkpoint.ipynb
Untitled5-checkpoint.ipynb
Untitled6-checkpoint.ipynb
Untitled7-checkpoint.ipynb
Untitled8-checkpoint.ipynb
Untitled9-checkpoint.ipynb
zoo-checkpoint.ipynb
history.sqlite
history.sqlite-journal
README
migrated
notebook.json
fontlist-v330.json
desktop.ini
.nonadmin
api-ms-win-core-console-l1-1-0.dll
api-ms-win-core-datetime-l1-1-0.dll
api-ms-win-core-debug-l1-1-0.dll
api-ms-win-core-errorhandling-l1-1-0.dll
api-ms-win-core-file-l1-1-0.dll
api-ms-win-core-file-l1-2-0.dll
api-ms-win-core-file-l2-1-0.dll
api-ms-win-core-handle-l1-1-0.dll
api-ms-win-core-heap-l1-1-0.dll
api-ms-win-core-interlocked-l1-1-0.dll
api-ms-win-core-libraryloader-l1-1-0.dll
api-ms-win-core-localization-l1-2-0.dll
api-ms-win-core-memory-l1-1-0.dll
api-ms-win-core-namedpipe-l1-1-0.dll
api-ms-win-core-processenvironment-l1-1-0.dll
api-ms-win-core-processthreads-l1-1-0.dll
api-ms-win-core-processthreads-l1-1-1.dll
api-ms-win-core-profile-l1-1-0.dll
api-ms-win-core-rtlsupport-l1-1-0.dll
api-ms-win-core-string-l1-1-0.dll
api-ms-win-core-synch-l1-1-0.dll
api-ms-win-core-synch-l1-2-0.dll
api-ms-win-core-sysinfo-l1-1-0.dll
api-ms-win-core-timezone-l1-1-0.dll
api-ms-win-core-util-l1-1-0.dll
api-ms-win-crt-conio-l1-1-0.dll
api-ms-win-crt-convert-l1-1-0.dll
api-ms-win-crt-environment-l1-1-0.dll
api-ms-win-crt-filesystem-l1-1-0.dll
api-ms-win-crt-heap-l1-1-0.dll
api-ms-win-crt-locale-l1-1-0.dll
api-ms-win-crt-math-l1-1-0.dll
api-ms-win-crt-multibyte-l1-1-0.dll
api-ms-win-crt-private-l1-1-0.dll
api-ms-win-crt-process-l1-1-0.dll
api-ms-win-crt-runtime-l1-1-0.dll
api-ms-win-crt-stdio-l1-1-0.dll
api-ms-win-crt-string-l1-1-0.dll
api-ms-win-crt-time-l1-1-0.dll
api-ms-win-crt-utility-l1-1-0.dll
concrt140.dll
cwp.py
LICENSE_PYTHON.txt
msvcp140.dll
msvcp140_1.dll
msvcp140_2.dll
msvcp140_codecvt_ids.dll
pyodbc.pyi
python.exe
python.pdb
python3.dll
python39.dll
python39.pdb
pythonw.exe
pythonw.pdb
qt.conf

5. Create a list of all of the files in your parent directory (minimum five files should be available).
Ans:-
    
import os 
os.listdir()

6. Use multiprocessing to create three separate processes. Make each one wait a random number of
seconds between one and five, print the current time, and then exit.
Ans:-
import multiprocessing
import time

def sleepy_man():
    print('Starting to sleep')
    time.sleep(1)
    print('Done sleeping')

tic = time.time()
p1 =  multiprocessing.Process(target= sleepy_man)
p2 =  multiprocessing.Process(target= sleepy_man)
p1.start()
p2.start()
toc = time.time()

print('Done in {:.4f} seconds'.format(toc-tic))

Output:-
    Done in 0.0130 seconds
    
7. Create a date object of your day of birth.
Ans:-
    
from datetime import datetime
my_dob = datetime.strptime('3/9/1997','%d/%m/%Y')
print(my_dob, type(my_dob))

Output:-
    1997-09-03 00:00:00 <class 'datetime.datetime'>

8. What day of the week was your day of birth?
Ans:-
from datetime import datetime
my_dob = datetime(1997,9,3)
my_dob.strftime("%A")

Output:-
    'Wednesday'
    
9. When will you be (or when were you) 10,000 days old?
Ans:-
    
from datetime import datetime, timedelta
my_dob = datetime.strptime("3/9/1997",'%d/%m/%Y')
future_date = my_dob-timedelta(10000)
future_date

Output:-
    datetime.datetime(1970, 4, 18, 0, 0)
    
```
